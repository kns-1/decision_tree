```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor
from matplotlib import pyplot as pp
```

```python
data = np.random.RandomState(1) #random dataset
X = np.sort(5 * data.rand(80, 1), axis=0) #converting into arrays
y = np.sin(X).ravel()
y[::5] += 3 * (0.5 - data.rand(16))
print("x= ",X)
print("y= ",y)
```

```python
regr_1 = DecisionTreeRegressor(max_depth=2) #decision tree regressor 
regr_2 = DecisionTreeRegressor(max_depth=5)
regr_1.fit(X, y)
regr_2.fit(X, y)
```

```python
X_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]
y_1 = regr_1.predict(X_test)
y_2 = regr_2.predict(X_test)
print("X_test= ",X_test)
print("y_1= ",y_1)
print("y_2= ",y_2)
```

```python
pp.figure()
pp.scatter(X, y, s=20, edgecolor="black",
            c="darkorange", label="data")
pp.plot(X_test, y_1, color="red",
         label="max_depth=2", linewidth=2)
pp.plot(X_test, y_2,label="max_depth=5", linewidth=2)
pp.xlabel("data")
pp.ylabel("target")
pp.title("Decision Tree Regression")
pp.legend()
pp.show()
```
